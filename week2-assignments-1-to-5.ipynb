#### MAE
The mean absolute error (MAE) is the simplest regression error metric to understand. Weâ€™ll calculate the residual for every data point, taking only the absolute value of each so that negative and positive residuals do not cancel out. We then take the average of all these residuals. Effectively, MAE describes the typical magnitude of the residuals. (MAE tells us how big of an error we can expect from the forecast on average.)

##### Note: I compared my function mas_func with the oputput from mean_absolute_error(x,x_hat), and they match. meaning my function is working. 

```python
x = [6,5,6,8]
x_hat = [3,5,9,3]
mean_absolute_error(x,x_hat)
```

```python
def mas_func(a,b):
 mae_sum = 0
 for x,y in zip(a,b):
    mae_sum += abs(x - y)
    print(mae_sum)
 mae = mae_sum/len(a)
 return mae
    
x = [6,5,6,8]
x_hat = [3,5,9,3]
mas_func(x, x_hat)
```

### MSE 
the difference between the actual and the predicted values, But squares the difference before summing them all instead of using the absolute value. 

##### note: I compared my function mse_func with the output from mean_squared_error(x,x_hat) and they match, meaning my function is working.

```python
x = [6,5,6,8]
x_hat = [3,5,9,3]
mean_squared_error(x,x_hat)
```

```python
def mse_func(a,b):
 mse_sum = 0
 for x , y in zip(a,b):
    mse_sum += abs(x - y)**2
    #print(mse_sum)
 mse = mse_sum/len(a)
 return mse
    
x = [6,5,6,8]
x_hat = [3,5,9,3]
mse_func(x, x_hat)
```
